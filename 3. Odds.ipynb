```python
import os, json
import pandas as pd
import numpy as np
```

```python
years = [year for year in range(2016,2022+1)]
odds_dfs = []
for year in years:
    path_str = 'data/Odds/' + str(year) + '/'
    for x, file_name in enumerate([x_json for x_json in os.listdir(path_str) if x_json.endswith('.json')]):
        with open(path_str + file_name) as json_file:
            if x == 0:
                odds_df = pd.DataFrame(json.load(json_file))
            else:
                odds_df = pd.concat([odds_df,pd.DataFrame(json.load(json_file))])
    odds_df['Season'] = year
    odds_dfs.append(odds_df)
```

```python
full_odds = pd.concat(odds_dfs).reset_index(drop = True)
```

```python
full_odds
```

```python
full_odds['home_team'] = np.where(full_odds['Favorite'].str[:2]=="at",full_odds['Favorite'].str[3:],full_odds['UnderDog'].str[3:])
full_odds['away_team'] = np.where(full_odds['Favorite'].str[:2]=="at",full_odds['UnderDog'],full_odds['Favorite'])
full_odds['home_fav'] = np.where(full_odds['Favorite'].str[:2]=="at",True,False)
```

```python
full_odds['date_match'] = pd.to_datetime(full_odds['Date']).dt.normalize()
```

```python
full_odds
```

```python
names_key = ['Broncos', 'Texans', 'Jets', 'Eagles', 'Jaguars', 'Chiefs',
       'Saints', 'Titans', 'Ravens', 'Falcons', 'Seahawks', 'Colts',
       'Cowboys', 'Cardinals', 'Commanders', '49ers', 'Buccaneers',
       'Panthers', 'Chargers', 'Steelers', 'Patriots', 'Giants',
       'Lions', 'Vikings', 'Bengals', 'Browns', 'Rams', 'Raiders',
       'Dolphins', 'Bears', 'Bills', 'Packers']
```

```python
abbr_val = ['DEN','HOU','NYJ','PHI','JAX','KC','NO','TEN','BAL','ATL','SEA','IND','DAL','ARI','WAS','SF','TB','CAR','LAC',
           'PIT','NE','NYG','DET','MIN','CIN','CLE','LAR','LV','MIA','CHI','BUF','GB',]
```

```python
odds_to_abbr_map = {names_key[i]:abbr_val[i] for i in range(len(names_key))}
```

```python
full_odds['home_abbr'] = full_odds['home_team'].map(odds_to_abbr_map)
```

```python
full_odds['away_abbr'] = full_odds['away_team'].map(odds_to_abbr_map)
```

```python
full_odds['AwayTeamMoneyLine'] = pd.to_numeric(full_odds.AwayTeamMoneyLine)
full_odds['HomeTeamMoneyLine'] = pd.to_numeric(full_odds.HomeTeamMoneyLine)
```
