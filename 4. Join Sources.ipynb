```python
import os, json
import pandas as pd
import numpy as np
```

```python
games = pd.read_csv('data/game_results.csv',index_col=0)
odds = pd.read_csv('data/odds_final.csv',index_col=0)
salaries = pd.read_csv('data/dfs_salaries.csv',index_col=0)
```

```python
games_odds = pd.merge(games, odds, how = 'left', left_on=['Date','home_abbr','away_abbr'],right_on=['date_match','home_abbr','away_abbr'])
```

```python
pd.set_option('display.max_columns', None)
```

```python
games_odds
```

```python
games_odds.groupby(['season'])['season'].count()
```

```python
salary_home = pd.merge(salaries, games[['home_abbr','season','Week']], how = 'left', left_on=['Team','Season','Week'],right_on=['home_abbr','season','Week'])
```

```python
salary_home['home_team'] = salary_home['home_abbr'].notna()
```

```python
salary_home['homeaway'] = np.where(salary_home['home_team'], "home", "away")
```

```python
salary_home['key'] = np.where(salary_home['home_team'], salary_home['Season'].astype(str) + salary_home['Week'].astype(str) +salary_home['Team'],salary_home['Season'].astype(str) + salary_home['Week'].astype(str) +salary_home['Opponent'])
```

```python
salary_home
```

```python
home_away_salaries = salary_home.pivot(index = ['key'],columns = 'homeaway',values = ['DST1','QB1','RB1','RB2','WR1','WR2','WR3','WR4','TE1','TE2']).reset_index()
```

```python
home_away_salaries
```

```python
home_away_salaries.columns = ['_'.join(col).strip() for col in home_away_salaries.columns.values]
```

```python
home_away_salaries
```

```python
games_odds['key'] = games_odds['season'].astype(str) + games_odds['Week'].astype(str) +games_odds['home_abbr']
```

```python
full_games_w_sals = pd.merge(games_odds, home_away_salaries, how = 'left', left_on=['key'],right_on=['key_'])
```

```python
full_games_w_sals.drop(['home_team_x','away_team_x','Date_x','Date_y','Season','Favorite','UnderDog','home_team_y','away_team_y','key','key_'],axis = 1, inplace = True)
```

```python
full_games_w_sals.to_csv('data/full_data.csv')
```
