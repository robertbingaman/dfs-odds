```python
import pandas as pd
import numpy as np
import time
import random
```

```python
current_week = 13
current_season = 2022
```

```python
season_dfs = []
for year in range(2016,2023):
    df_temp = pd.read_html("https://www.pro-football-reference.com/years/"+str(year)+"/games.htm")
    df_temp[0]['season']=year
    season_dfs.append(df_temp[0])
    time.sleep(random.uniform(60,90))
```

```python
season_dfs[6]
```

```python
games_df = pd.concat(season_dfs)
```

```python
games_df.to_csv('data/games_full.csv')
```

```python
games_df.reset_index(drop = True, inplace = True)
```

```python
games_df['Unnamed: 5'] == '@'
```

```python
games_df['home_team'] = np.where(games_df['Unnamed: 5'] == '@',games_df['Loser/tie'],games_df['Winner/tie'])
games_df['away_team'] = np.where(games_df['Unnamed: 5'] == '@',games_df['Winner/tie'],games_df['Loser/tie'])
```

```python
games_df['tie'] = np.where(games_df['PtsW'] == games_df['PtsL'],True,False)
```

```python
games_df['home_win'] = np.where(games_df['Winner/tie'] == games_df['home_team'],1,0)
```

```python
games_df['home_pts'] = np.where(games_df['home_win'] == 1, games_df['PtsW'], games_df['PtsL'])
games_df['away_pts'] = np.where(games_df['home_win'] == 0, games_df['PtsW'], games_df['PtsL'])
```

```python
games_df
```

```python
games_df.drop(games_df[games_df['Week'] == 'Week'].index)
```

```python
games_df.Week.unique()
```

```python
drop_vals=['Week',np.nan,'WildCard','Division','ConfChamp','SuperBowl']
```

```python
for x in drop_vals:
    games_df.drop(games_df[games_df['Week'] == x].index, inplace = True)
    games_df.reset_index(drop = True, inplace = True)
```

```python
games_df['home_pts'] = pd.to_numeric(games_df['home_pts'])
games_df['away_pts'] = pd.to_numeric(games_df['away_pts'])
```

```python
games_df['total_pts']  = games_df['home_pts'] + games_df['away_pts']
```

```python
games_df_final = games_df[['season','Week','home_team','away_team','home_pts','away_pts','total_pts','home_win','Day','Date','Time']]
```

```python
team_names = ['Denver Broncos', 'Baltimore Ravens', 'Jacksonville Jaguars',
       'Atlanta Falcons', 'Houston Texans', 'New York Jets',
       'Philadelphia Eagles', 'Kansas City Chiefs', 'Tennessee Titans',
       'New Orleans Saints', 'Seattle Seahawks', 'Indianapolis Colts',
       'Dallas Cowboys', 'Arizona Cardinals', 'Washington Redskins',
       'San Francisco 49ers', 'Buffalo Bills', 'Pittsburgh Steelers',
       'Cleveland Browns', 'Detroit Lions', 'New England Patriots',
       'Carolina Panthers', 'New York Giants', 'Los Angeles Rams',
       'Oakland Raiders', 'San Diego Chargers', 'Minnesota Vikings',
       'Chicago Bears', 'Cincinnati Bengals', 'Miami Dolphins',
       'Green Bay Packers', 'Tampa Bay Buccaneers',
       'Los Angeles Chargers', 'Washington Football Team',
       'Las Vegas Raiders', 'Washington Commanders']
abbr_val = ['DEN','BAL','JAX','ATL','HOU','NYJ','PHI','KC','TEN','NO','SEA','IND','DAL','ARI','WAS','SF','BUF','PIT','CLE',
            'DET','NE','CAR','NYG','LAR','LV','LAC','MIN','CHI','CIN','MIA','GB','TB','LAC','WAS','LV','WAS']
game_name_abbr_map = {team_names[i]:abbr_val[i] for i in range(len(team_names))}
```

```python
games_df_final['home_abbr'] = games_df_final['home_team'].map(game_name_abbr_map)
games_df_final['away_abbr'] = games_df_final['away_team'].map(game_name_abbr_map)
```

```python
games_df_train = games_df_final.dropna(subset = ['home_pts'])
games_df_train.reset_index(drop = True, inplace = True)
```

```python
games_df_train.to_csv('data/game_results.csv')
```
